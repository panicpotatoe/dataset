```python
import pandas as pd

df = pd.read_csv("../bird.csv")
df.sample(5)
```

```python
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.pie(x=df["no"].value_counts(), labels=df["species"])
plt.show()
```

```python
COLOR_CODE = ["red", "green", "blue"]
```

```python
# create pivot table with 2 columns: species, count of each species
df_pivot = df.pivot_table(index="species", values="no", aggfunc="count")
df_pivot
```

```python
explode = (0.5, 0.5, 0.5)
plt.pie(
    x=df_pivot["no"].values,
    labels=df_pivot.index,  # index là tên cột đầu tiên, không phải "spieces" đâu
    autopct="%1.1f%%",  # định dạng phần trăm
    startangle=90,  # góc bắt đầu
    shadow=True,  # có bóng
    colors=COLOR_CODE,  # màu sắc
    explode=explode,  # làm nổi bật phần tử thứ 1
)
# add title
plt.title("Bird Species Distribution")

plt.show()
```
