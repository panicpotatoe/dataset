```python
import pandas as pd

df = pd.read_csv("../bird.csv")
df.sample(5)
```

```python
import matplotlib.pyplot as plt

plt.pie(x=df["no"].value_counts(), labels=df["species"])
plt.show()
```

```python
COLOR_CODE = ["red", "green", "blue"]
```

```python
# create pivot table with 2 columns: species, count of each species
df_pivot = df.pivot_table(index="species", values="no", aggfunc="count")
df_pivot
```

```python
explode = (0.5, 0.5, 0.5)
plt.pie(
    x=df_pivot["no"].values,
    labels=df_pivot.index,  # index lÃ  tÃªn cá»™t Ä‘áº§u tiÃªn, khÃ´ng pháº£i "spieces" Ä‘Ã¢u
    autopct="%1.1f%%",  # Ä‘á»‹nh dáº¡ng pháº§n trÄƒm
    startangle=90,  # gÃ³c báº¯t Ä‘áº§u
    shadow=True,  # cÃ³ bÃ³ng
    colors=COLOR_CODE,  # mÃ u sáº¯c
    explode=explode,  # lÃ m ná»•i báº­t pháº§n tá»­ thá»© 1
)
# add title
plt.title("Bird Species Distribution")

plt.show()
```
